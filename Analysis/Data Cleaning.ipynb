## 2. Data Cleaning

### 2-(1) Datatype Conversion

```python
order_list.info()
```

```python
# '호출일시' datetime 이상치 제거
order_list = order_list.iloc[1:-2]
```

```python
# Convert Datatype for columns : 호출일시, 탑승일시, 하차일시, 배차성공일시, 취소일시 

def convert(x):
    x = x.str.split('.').str[0].astype('datetime64')
    return x

order_list['call_time'] = convert(order_list['호출일시'])
order_list['board_time'] = convert(order_list['탑승일시'])
order_list['alight_time'] = convert(order_list['하차일시'])
order_list['suc_time'] = convert(order_list['배차성공일시'])
order_list['cancel_time'] = convert(order_list['취소일시'])
```

```python
order_list.info()
```

```python
order_list = order_list.drop(['호출일시', '탑승일시', '하차일시', '배차성공일시', '취소일시'], axis = 1)
```

```python
# Split latitude and longitude for column '호출장소'

order_list[["x", "y"]] = order_list["호출장소"].str.strip(r"()").str.split(",", expand = True).astype(float)

order_list = order_list.drop(['호출장소'], axis = 1)
```

```python
order_list.info()
```

### 2-(2) Rename columns / Unit

```python
order_list = order_list.rename(columns = {"이용자ID":"ID", "탑승정류장":"board_station", "하차정류장":"alight_station", "어른(명)":"adult", "청소년(명)":"teen", "어린이(명)":"child", 
                                "이동거리(m)":"distance", "요금(원)":"fare", "배차상태":"suc_status", "배차상태상세":"suc_status_d"})
```

```python
# change unit of "distance" (m to km)

order_list['distance'] = order_list['distance'] / 1000
```

### 3-(2) 탑승희망 정류장의 변화 여부 관찰

```python
# Trip Chain

order_sort['Trip'] = 0

# Check if the alighting station is different from the station of previous row

# Function to update the 'Trip' column based on the conditions
def update_trip(row):
    global prev_trip,prev_station
    if row['board_station'] != prev_station:
        prev_trip += 1
    else:
        pass
    prev_station = row['board_station']
    return prev_trip

# Set the initial value for prev_trip
prev_trip = 0
prev_station = order_sort.at[0, 'board_station']

# Update the 'Trip' column using the apply function
order_sort['Trip'] = order_sort.apply(update_trip, axis=1)

```
