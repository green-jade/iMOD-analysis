## 정류장 plotting

```python
import geopandas
import folium
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/st_no.csv')
df.head()
```

```python
# keep only relevant columns
df = df.loc[:,("id","longitude","latitude")]
df.head()
```

```python
# convert xy to geometry points
geometry = geopandas.points_from_xy(df.longitude, df.latitude)
geo_df = geopandas.GeoDataFrame(df[['id','longitude','latitude']], geometry=geometry)
geo_df.head()
```

```python
import folium
map = folium.Map(location=[37.45, 126.5], zoom_start=12)
map
```

```python
geo_df_list = [[point.xy[1][0], point.xy[0][0]] for point in geo_df.geometry ]
for coordinates in geo_df_list:
    map.add_child(folium.Marker(location = coordinates, icon = folium.Icon(color = 'red')))
```
